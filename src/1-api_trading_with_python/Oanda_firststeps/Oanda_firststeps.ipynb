
### Practice Example with 5 min delay

#### 1. Initiate Constants

```python
INSTRUMENT = "EUR_USD"
UNITS = 100000
DELAY = 5*60 # 5 minutes
```

#### 2. Long order

```python

open = api.create_order(instrument = INSTRUMENT, units = UNITS, sl_distance= 0.1, ret = True)
```

#### 3. Wait for timer + Close postiion

```python
import time

time.sleep(DELAY)
close = api.create_order(instrument = INSTRUMENT, units = -UNITS, ret = True)
```

#### 4. Analyze trade

```python
print(open['id'])
print(close['id'])
transaction = api.get_transactions(tid = open['id'])
print("Profit/Loss: ", transaction[2]["pl"])

```
